```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler



# load data
train_data = pd.read_csv('/kaggle/input/playground-series-s3e20/train.csv')

# Cek data 'train_data'
print(train_data.head())

# Periksa struktur data, jenis data, dan apakah ada nilai yang hilang :
# info dataset
print(train_data.info())
# Statistik deskriptif
print(train_data.describe())
# Cek nilai yang hilang
print(train_data.isnull().sum())

# Mengisi nilai yang hilang dengan rata-rata
train_data.fillna(train_data .mean(), inplace=True)

# Lakukan normalisasi atau transformasi pada fitur numerik
# Normalisasi
scaler = MinMaxScaler()
train_data[['SulphurDioxide_SO2_column_number_density', 'SulphurDioxide_SO2_column_number_density_amf', '...other_numerik_features...']] = scaler.fit_transform(train_data[['SulphurDioxide_SO2_column_number_density', 'SulphurDioxide_SO2_column_number_density_amf', '...oher_numeric_features...']])

# Buat time series plot untuk melihat tren dan pola
# Misalnya, untuk lokasi tertentu
location_data = train_data[train_data['ID_LAT_LON_YEAR_WEEK'] == 'ID_-0.510_29.290_2019_00']

# Plot time series
plt.figure(figsize=(12, 6))
plt.plot(location_data['week_no'], location_data['SulphurDioxide_SO2_column_number_density'], label='SulphurDioxide_SO2_column_number_density')
# ... Plot fitur lainnya ...
plt.xlabel('Week')
plt.ylabel('Value')
plt.title('Time Series Plot for Location ID_-0.510_29.290_2019_00')
plt.legend()
plt.show()

column_to_normalize = [
    'SulphurDioxide_SO2_column_number_density',
    'SulphurDioxide_SO2_column_number_density_amf',
    'SulphurDioxide_SO2_slant_column_number_density',
    'SulphurDioxide_cloud_fraction',
    'SulphurDioxide_sensor_azimuth_angle',
]

scaler = MinMaxScaler()
trasin_data[column_to_normalize] = scaler.fit_transform(train_data[column_to_normalize])
print(train_data.head())
```
